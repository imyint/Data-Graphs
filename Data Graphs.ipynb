```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm

# create data
names=['White', 'Asian', 'Black or African American', 'Hispanic/Latino', 'Multiracial', 'Other']
size=[139, 58, 19, 18, 11, 7]
total_size = sum(size[0:len(size)])
percentages = (100*x/total_size for x in size)


# circle for center of plot
middle_circle = plt.Circle( (0,0), 0.6, color='white')

# plt.figure(facecolor = 'lightgray')


# title and color names
title = plt.title('Ethnic Diversity', fontsize = 35)
title.set_ha("left")
patches, texts, autotexts = plt.pie(size, colors=['red','green','blue','skyblue', 'yellow', 'orange'], autopct='%.2f%%', pctdistance =0.8, labeldistance=0.75)

for autotext in autotexts:
    autotext.set_color('black')

# label font size
proptease = fm.FontProperties()
proptease.set_size('xx-large')
plt.setp(autotexts, fontproperties=proptease)
plt.setp(texts, fontproperties=proptease)

p=plt.gcf()
p.gca().add_artist(middle_circle)
p.set_size_inches(10,10)
plt.axis('equal')

# legend position
plt.legend(labels = names,bbox_to_anchor=(1,0.62), loc="center right", fontsize=15, borderaxespad=3, 
           bbox_transform=plt.gcf().transFigure )
plt.subplots_adjust(left=0, bottom=0.3, right=0.55)


plt.show()
```
